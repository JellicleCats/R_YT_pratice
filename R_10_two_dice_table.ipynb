```r
# sum of one dice

roll <- 100
dice1 <- ceiling(runif(roll)*6)
table(dice1)
```

```r
# sum of two dices

roll <- 100
dice2 <- ceiling(runif(roll)*6)+ceiling(runif(roll)*6)
table(dice2)
```

```r
# sum of two dices

roll <- 100
dice2 <- ceiling(runif(roll)*6)+ceiling(runif(roll)*6)
a <- table(dice2)
barplot(a)
```
